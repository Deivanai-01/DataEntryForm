```python
from tkinter import *
from tkinter.ttk import Combobox
import tkinter as tk
from tkinter import messagebox
import openpyxl ,xlrd
from openpyxl import workbook
import pathlib

root=Tk()
root.title("Employee Registration From")
root.geometry('800x500+300+200')
root.resizable(True,True)
root.configure(bg="#326273")

file=pathlib.Path('Employee_Registration.xlsx')
if file.exists():
    pass
else:
    file=openpyxl.Workbook()
    sheet=file.active
    sheet['A1']="Employee ID"
    sheet['B1']="Employee Name"
    sheet['C1']="Date of Birth"
    sheet['D1']="Age"
    sheet['E1']="Gender"
    sheet['F1']="Qualification"
    sheet['G1']="Moblie Number"
    sheet['H1']="Email ID"
    sheet['I1']="Address"

    file.save('Employee_Registration.xlsx')

def submit():
    id=idEntry.get()
    name=nameEntry.get()
    dob=dobEntry.get()
    age=ageEntry.get()
    gender=gender_combobox.get()
    qua=quaEntry.get()
    num=noEntry.get()
    email=emailidEntry.get()
    add=addressEntry.get()
    
    file=openpyxl.load_workbook('Employee_Registration.xlsx')
    sheet=file.active
    sheet.cell(column=1,row=sheet.max_row+1,value=id)
    sheet.cell(column=2,row=sheet.max_row,value=name)
    sheet.cell(column=3,row=sheet.max_row,value=dob)
    sheet.cell(column=4,row=sheet.max_row,value=age)
    sheet.cell(column=5,row=sheet.max_row,value=gender)
    sheet.cell(column=6,row=sheet.max_row,value=qua)
    sheet.cell(column=7,row=sheet.max_row,value=num)
    sheet.cell(column=8,row=sheet.max_row,value=email)
    sheet.cell(column=9,row=sheet.max_row,value=add)
    
    file.save(r'Employee_Registration.xlsx')
    
    messagebox.showinfo('info','Employee Details added')
    
    idEntry.delete(0,"end")
    nameEntry.delete(0,"end")
    dobEntry.delete(0,"end")
    ageEntry.delete(0,"end")
    quaEntry.delete(0,"end")
    noEntry.delete(0,"end")
    emailidEntry.delete(0,"end")
    addressEntry.delete(0,"end")
    gender_combobox.set('')

    
def clear():
    idEntry.delete(0,"end")
    nameEntry.delete(0,"end")
    dobEntry.delete(0,"end")
    ageEntry.delete(0,"end")
    quaEntry.delete(0,"end")
    noEntry.delete(0,"end")
    emailidEntry.delete(0,"end")
    addressEntry.delete(0,"end")
    gender_combobox.set('')

#Label

Label(root,text="Employee ID",font=23,bg="#326273",fg="#fff").place(x=50,y=50)
Label(root,text="Employee Name",font=23,bg="#326273",fg="#fff").place(x=50,y=100)
Label(root,text="Date of Birth",font=23,bg="#326273",fg="#fff").place(x=50,y=150)
Label(root,text="Age",font=23,bg="#326273",fg="#fff").place(x=370,y=150)
Label(root,text="Gender",font=23,bg="#326273",fg="#fff").place(x=50,y=200)
Label(root,text="Qualification",font=23,bg="#326273",fg="#fff").place(x=50,y=250)
Label(root,text="Moblie Number",font=23,bg="#326273",fg="#fff").place(x=50,y=300)
Label(root,text="Email ID",font=23,bg="#326273",fg="#fff").place(x=50,y=350)
Label(root,text="Address",font=23,bg="#326273",fg="#fff").place(x=50,y=400)

#entry

idEntry=Entry(root,textvariable=StringVar(),width=40,bd=2,font=20)
nameEntry=Entry(root,textvariable=StringVar(),width=40,bd=2,font=20)
dobEntry=Entry(root,textvariable=StringVar(),width=15,bd=2,font=20)
ageEntry=Entry(root,textvariable=StringVar(),width=15,bd=2,font=20)
gender_combobox=Combobox(root,values=['Male','Female'],font='arial 14',state='r',width=31)
quaEntry=Entry(root,textvariable=StringVar(),width=40,bd=2,font=20)
noEntry=Entry(root,textvariable=StringVar(),width=40,bd=2,font=20)
emailidEntry=Entry(root,textvariable=StringVar(),width=40,bd=2,font=20)
addressEntry=Entry(root,textvariable=StringVar(),width=40,bd=2,font=20)


idEntry.place(x=200,y=50)
nameEntry.place(x=200,y=100)
dobEntry.place(x=200,y=150)
ageEntry.place(x=420,y=150)
gender_combobox.place(x=200,y=200)
quaEntry.place(x=200,y=250)
noEntry.place(x=200,y=300)
emailidEntry.place(x=200,y=350)
addressEntry.place(x=200,y=400)

#Button

Button(root,text='Submit',bg="#326273",fg="white",width=15,height=2,command=submit).place(x=200,y=450)
Button(root,text='Clear',bg="#326273",fg="white",width=15,height=2,command=clear).place(x=340,y=450)
Button(root,text='Exit',bg="#326273",fg="white",width=15,height=2,command=lambda:root.destroy()).place(x=480,y=450)


root.mainloop()
```
